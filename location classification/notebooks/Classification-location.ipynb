```python
oe=OrdinalEncoder()
le=LabelEncoder()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
descriptions=tfidf.fit_transform(df['Description'])
```

```python
df2=pd.DataFrame(descriptions.toarray(),columns=tfidf.get_feature_names_out())
```

```python
df2
```

```python
# df[["Activity","Type"]]=oe.fit_transform(df[["Activity","Type"]])
df[["Description"]]=oe.fit_transform(df[["Description"]])
df[["Price"]]=oe.fit_transform(df[["Price"]])
```

```python
df
```

```python
X=df2
y=df["location"]
```

```python
overSample=RandomOverSampler()
```

```python
X_reSample,y_reSample=overSample.fit_resample(X,y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_reSample,y_reSample,test_size=0.2)
```

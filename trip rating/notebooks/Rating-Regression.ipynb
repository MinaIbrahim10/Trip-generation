```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from imblearn.over_sampling import RandomOverSampler,SMOTE
```

```python
df=pd.read_csv("trip_data_with_rating.csv")
df
```

```python
le=LabelEncoder()
oe=OrdinalEncoder()
tfidf=TfidfVectorizer()
```

```python
df["location"]=le.fit_transform(df["location"])
df["Price"]=le.fit_transform(df["Price"])
df
```

```python
activities=tfidf.fit_transform(df["Activity"])
```

```python
df2=pd.DataFrame(activities.toarray(),columns=tfidf.get_feature_names_out())
```

```python
df2
```

```python
df3=pd.concat([df[["location","Price"]],df2],axis=1)
```

```python
df3
```

```python
X=df3
y=df["rating"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
DT=DecisionTreeRegressor(max_depth=35)
```

```python
DT.fit(X_train,y_train)
```

```python
DT.score(X_train,y_train)
```

```python
DT.score(X_test,y_test)
```

```python
RF=RandomForestRegressor()
```

```python
RF.fit(X_train,y_train)
```

```python
RF.score(X_train,y_train)
```

```python
RF.score(X_test,y_test)
```

```python
EXT=ExtraTreesRegressor()
```

```python
EXT.fit(X_train,y_train)
```

```python
EXT.score(X_train,y_train)
```

```python
EXT.score(X_test,y_test)
```

```python
RF.predict(X_test[1:2])
```

```python
y_test[1:2]
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_percentage_error,r2_score,max_error,explained_variance_score
```

```python
y_predict=RF.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_predict)
mae=mean_absolute_percentage_error(y_test,y_predict)
r2=r2_score(y_test,y_predict)
max_e=max_error(y_test,y_predict)
evs=explained_variance_score(y_test,y_predict)
```

```python
print(mse)
print(mae)
print(r2)
print(max_e)
print(evs)
```
